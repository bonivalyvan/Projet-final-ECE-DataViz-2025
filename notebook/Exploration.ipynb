```python
import requests
import zipfile
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.ticker as ticker
```

```python
# URL du fichier
url = "https://archive.ics.uci.edu/static/public/502/online+retail+ii.zip"

# T√©l√©charger le fichier zip
print("T√©l√©chargement en cours...")
response = requests.get(url)
response.raise_for_status()  # V√©rifie si le t√©l√©chargement s'est bien pass√©

# Ouvrir le zip en m√©moire
print("Extraction et lecture...")
with zipfile.ZipFile(io.BytesIO(response.content)) as file:
    # Liste des fichiers dans le zip
    print("Fichiers dans le zip :", file.namelist())
    
    # On sait que le fichier s'appelle 'online_retail_II.xlsx'
    file_name = file.namelist()[0]
    sheets_dict = pd.read_excel(file.open(file_name), engine='openpyxl', sheet_name=None)
    print("Feuilles trouv√©es :", sheets_dict.keys())

    # Lire le fichier txt directement dans un DataFrame
    df = pd.concat(sheets_dict.values(), ignore_index=True)
    #df = pd.read_excel(file.open(file_name), engine='openpyxl')

    #afficher du DataFrame
    print(f"Nombre total de lignes : {len(df)}")
    print(df.head())
```

Nettoyage des donn√©es manquantes et de valeurs ab√©rrantes afin d'obtenir des graphique fid√®les √† la r√©alit√© observer par les ventes

```python
# Cr√©er une copie pour le nettoyage (pr√©server df original)
df_clean = df.copy()

print('√âtat initial avant nettoyage:')
print(f'Nombre de lignes: {len(df_clean)}')
print(f'Nombre de colonnes: {len(df_clean.columns)}')

#Supprimer les doublons identiques
initial_rows = len(df_clean)
df_clean = df_clean.drop_duplicates()
removed_dups = initial_rows - len(df_clean)
print(f'\nDoublons exacts supprim√©s: {removed_dups} lignes')

#Supprimer les lignes avec CustomerID manquant (essentiel pour analyse)
initial_rows = len(df_clean)
df_clean = df_clean[df_clean['Customer ID'].notna()]
removed_missing_customer = initial_rows - len(df_clean)
print(f'Lignes sans CustomerID supprim√©es: {removed_missing_customer} lignes')

#Ajouter colonne Total (Quantity * Price) pour validation
df_clean['Total'] = df_clean['Quantity'] * df_clean['Price']

#on gardes de cot√© toutes les commandes annul√©e pour analyse ult√©rieure
df_clean['Invoice'] = df_clean['Invoice'].astype(str)
mask_returns = df_clean['Invoice'].str.startswith('C')
df_returns = df_clean[mask_returns].copy()
df_returns['Quantity'] = df_returns['Quantity'].abs()
df_returns['Price'] = df_returns['Price'].abs()
df_returns['Total'] = df_returns['Total'].abs()

#Quantit√©: supprimer les valeurs <= 0 (annulations/erreurs) le reste dans df_clean
initial_rows = len(df_clean)
df_clean = df_clean[df_clean['Quantity'] > 0]
removed_qty_zero = initial_rows - len(df_clean)
print(f'Retours isol√©s dans df_returns : {len(df_returns)} lignes')
print(f'Lignes avec Quantity ‚â§ 0 supprim√©es(df_clean): {removed_qty_zero} lignes')

#Prix: supprimer les valeurs <= 0 (erreurs)
initial_rows = len(df_clean)
df_clean = df_clean[df_clean['Price'] > 0]
removed_price_zero = initial_rows - len(df_clean)
print(f'Lignes avec Price ‚â§ 0 supprim√©es: {removed_price_zero} lignes')

#Supprimer les dates invalides (apr√®s conversion en datetime)
initial_rows = len(df_clean)
df_clean = df_clean[df_clean['InvoiceDate'].notna()]
removed_invalid_date = initial_rows - len(df_clean)
print(f'Lignes avec date invalide supprim√©es: {removed_invalid_date} lignes')

#Supprimer les lignes avec Description manquante
initial_rows = len(df_clean)
df_clean = df_clean[df_clean['Description'].notna()]
removed_missing_desc = initial_rows - len(df_clean)
print(f'Lignes sans Description supprim√©es: {removed_missing_desc} lignes')


#R√©initialiser l'index apr√®s tous les filtres
df_clean = df_clean.reset_index(drop=True)

print(f'\nNETTOYAGE TERMIN√â')
print(f'\n√âtat final apr√®s nettoyage:')
print(f'Nombre de lignes: {len(df_clean)}')
print(f'Nombre de colonnes: {len(df_clean.columns)}')
print(f'% de donn√©es conserv√©es: {(len(df_clean) / len(df)) * 100:.1f}%')

# Afficher les statistiques des colonnes num√©riques apr√®s nettoyage
print(f'\nStatistiques apr√®s nettoyage:')
print(df_clean[['Quantity', 'Price', 'Total']].describe())

# Afficher les colonnes manquantes (devrait √™tre minimal)
print(f'\nValeurs manquantes restantes:')
missing_clean = df_clean.isna().sum()
missing_clean = missing_clean[missing_clean > 0]
if len(missing_clean) > 0:
    display(missing_clean)
else:
    print('Aucune valeur manquante dans les colonnes critiques ')
    
```

Suppression des Outliers (C'est un intrus, une anomalie statistique qui se situe tr√®s loin de la "normale".)


```python
#Outliers sur Price
Q1_price = df_clean['Price'].quantile(0.25)
Q3_price = df_clean['Price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_price = Q1_price - 1.5 * IQR_price
upper_price = Q3_price + 1.5 * IQR_price

initial_rows = len(df_clean)
df_clean = df_clean[(df_clean['Price'] >= lower_price) & (df_clean['Price'] <= upper_price)]
removed_price_outliers = initial_rows - len(df_clean)
print(f'Outliers Price supprim√©s (IQR method): {removed_price_outliers} lignes')

# Outliers sur Quantity (ex: commandes aberrantes)
Q1_qty = df_clean['Quantity'].quantile(0.25)
Q3_qty = df_clean['Quantity'].quantile(0.75)
IQR_qty = Q3_qty - Q1_qty
lower_qty = Q1_qty - 1.5 * IQR_qty
upper_qty = Q3_qty + 1.5 * IQR_qty

initial_rows = len(df_clean)
df_clean = df_clean[(df_clean['Quantity'] >= lower_qty) & (df_clean['Quantity'] <= upper_qty)]
removed_qty_outliers = initial_rows - len(df_clean)
print(f'Outliers Quantity supprim√©s (IQR method): {removed_qty_outliers} lignes')

print(f'\nNETTOYAGE TERMIN√â')
print(f'\n√âtat final apr√®s nettoyage:')
print(f'Nombre de lignes: {len(df_clean)}')
print(f'Nombre de colonnes: {len(df_clean.columns)}')
print(f'% de donn√©es conserv√©es: {(len(df_clean) / len(df)) * 100:.1f}%')

# Afficher les statistiques des colonnes num√©riques apr√®s nettoyage
print(f'\nStatistiques apr√®s nettoyage:')
print(df_clean[['Quantity', 'Price', 'Total']].describe())
```

Graphique 1: √âvolution du chiffre d‚Äôaffaires mensuel

Type : line chart

```python
#Graphique 1: √âvolution du chiffre d‚Äôaffaires mensuel

# prepare
dfm = df_clean.copy()
dfm['Month'] = dfm['InvoiceDate'].dt.to_period('M').dt.to_timestamp()
sales_monthly = dfm.groupby('Month')['Total'].sum().reset_index()
sales_monthly['MA3'] = sales_monthly['Total'].rolling(3, center=True, min_periods=1).mean()

# plot
plt.figure(figsize=(12,5))
sns.lineplot(data=sales_monthly, x='Month', y='Total', marker='o', linewidth=2.2, label='CA mensuel')
sns.lineplot(data=sales_monthly, x='Month', y='MA3', linestyle='--', linewidth=2, color='gray', label='Moyenne mobile 3 mois')

# highlight peaks
peak = sales_monthly.loc[sales_monthly['Total'].idxmax()]
low = sales_monthly.loc[sales_monthly['Total'].idxmin()]
plt.scatter([peak['Month']], [peak['Total']], color='tab:red', s=100, zorder=5)
plt.scatter([low['Month']], [low['Total']], color='tab:orange', s=80, zorder=5)
plt.annotate(f"Max: ¬£{int(peak['Total']):,}", xy=(peak['Month'], peak['Total']), xytext=(0,15), textcoords='offset points', ha='center')
plt.annotate(f"Min: ¬£{int(low['Total']):,}", xy=(low['Month'], low['Total']), xytext=(0,-20), textcoords='offset points', ha='center')

# optional: shade holiday months (Nov-Dec)
import datetime as dt
for year in sorted(sales_monthly['Month'].dt.year.unique()):
    start = pd.Timestamp(year=year, month=11, day=1)
    end = pd.Timestamp(year=year, month=12, day=31)
    plt.axvspan(start, end, alpha=0.08, color='green')

plt.title("√âvolution mensuelle du chiffre d'affaires (avec trend & highlights)")
plt.ylabel("CA monthly (¬£)")
plt.xlabel("")
plt.xticks(rotation=45)
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()

```

5. Cohortes d‚Äôacquisition ‚Äî Base pour cohorte

Objectifs :

Comprendre la r√©tention mensuelle

Pr√©parer la matrice de cohortes

Graphique 5: Heatmap des cohortes

```python
# --- 1. Calcul de la Matrice de R√©tention ---
# Cr√©ation du mois de facture et du mois de cohorte (premi√®re acquisition)
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M')
df_clean['CohortMonth'] = df_clean.groupby('Customer ID')['InvoiceMonth'].transform('min')

# Calcul de l'index de cohorte (nombre de mois √©coul√©s)
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    return year, month

invoice_year, invoice_month = get_date_int(df_clean, 'InvoiceMonth')
cohort_year, cohort_month = get_date_int(df_clean, 'CohortMonth')
years_diff = invoice_year - cohort_year
months_diff = invoice_month - cohort_month

df_clean['CohortIndex'] = years_diff * 12 + months_diff + 1

# Cr√©ation de la table pivot
cohort_data = df_clean.groupby(['CohortMonth', 'CohortIndex'])['Customer ID'].nunique().reset_index()
cohort_counts = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='Customer ID')

# Calcul des pourcentages
cohort_sizes = cohort_counts.iloc[:, 0]
retention = cohort_counts.divide(cohort_sizes, axis=0)

# --- 2. Cr√©ation du Graphique Am√©lior√© ---
plt.figure(figsize=(16, 10))

# Formatage des labels de l'axe Y pour qu'ils soient lisibles (ex: "2010-01")
yticklabels = [x.strftime('%Y-%m') for x in retention.index]

# Heatmap
# vmin et vmax permettent de "zoomer" sur les couleurs int√©ressantes. 
# On met vmax √† 0.5 (50%) car le mois 1 est toujours √† 100% et √©crase les nuances sinon.
sns.heatmap(retention, 
            annot=True, 
            fmt='.0%', # Format pourcentage sans d√©cimale
            cmap='RdYlGn', # Rouge (faible) √† Vert (fort)
            vmin=0.0, 
            vmax=0.5, 
            linewidths=1, # Espace blanc entre les cases
            yticklabels=yticklabels)

plt.title('Analyse de R√©tention par Cohorte (en % des clients actifs)', fontsize=18, fontweight='bold', pad=20)
plt.ylabel('Mois d\'acquisition (Cohorte)', fontsize=14)
plt.xlabel('Mois √©coul√©s depuis le 1er achat', fontsize=14)
plt.yticks(rotation=0) # Mettre les dates √† l'horizontale

plt.show()
```

```python
# --- Pr√©paration des donn√©es ---
# Extraction du jour de la semaine et de l'heure
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.day_name()
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour

# On d√©finit l'ordre des jours (sinon ce sera alphab√©tique : Friday, Monday...)
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Sunday'] 
# Note : Il n'y a souvent pas de ventes le Samedi dans ce dataset (maintenance), √† v√©rifier.

# Cr√©ation de la table pivot (Nombre de commandes par Jour/Heure)
timing_data = df_clean.groupby(['DayOfWeek', 'Hour'])['Invoice'].nunique().reset_index()
timing_pivot = timing_data.pivot(index='DayOfWeek', columns='Hour', values='Invoice')

# R√©organiser l'index pour respecter l'ordre des jours
timing_pivot = timing_pivot.reindex(days_order)

# --- Graphique ---
plt.figure(figsize=(14, 7))
sns.heatmap(timing_pivot, cmap='coolwarm', annot=False, fmt='g', linewidths=0.5)

plt.title('Heatmap des Ventes : Jours vs Heures', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Heure de la journ√©e', fontsize=12)
plt.ylabel('Jour de la semaine', fontsize=12)
plt.yticks(rotation=0)
plt.show()
```

8 - Taux de retour mensuel
Ce graphique √† double axe r√©pond √† trois questions pr√©cises de ton √©nonc√© ("Quantifier effet sur marge et r√©tention")

Objectifs: 

Surveiller la d√©rive (La ligne rouge)

Identifier la saisonnalit√© (Les pics)

D√©tecter les incidents (Les anomalies)



```python
# --- Pr√©paration des donn√©es temporelles ---
# On cr√©e la colonne 'InvoiceMonth' pour les deux dataframes
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M')
df_returns['InvoiceMonth'] = df_returns['InvoiceDate'].dt.to_period('M')

# Aggr√©gation par mois
monthly_sales = df_clean.groupby('InvoiceMonth')['Total'].sum()
monthly_returns = df_returns.groupby('InvoiceMonth')['Total'].sum()

# Fusion des donn√©es (Sales + Returns)
return_analysis = pd.DataFrame({
    'Sales': monthly_sales,
    'Returns': monthly_returns
}).fillna(0) # Remplace les NaN par 0 (mois sans retours)

# Calcul du taux (%)
return_analysis['ReturnRate'] = (return_analysis['Returns'] / return_analysis['Sales']) * 100

# Cr√©ation du Graphique
fig, ax1 = plt.subplots(figsize=(14, 7))

# 1. Barres : Montant des Retours (Axe Gauche - Gris)
sns.barplot(
    x=return_analysis.index.astype(str), 
    y=return_analysis['Returns'], 
    ax=ax1, 
    color='lightgrey', 
    alpha=0.8, 
    label='Montant Rembours√© (¬£)'
)

# Config Axe Gauche
ax1.set_ylabel('Montant Rembours√© (¬£)', fontsize=12)
ax1.set_xlabel('Mois', fontsize=12)
ax1.tick_params(axis='x', rotation=45)
# Formatage nombres (ex: 10,000)
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y)))

#Ligne : Taux de Retour (Axe Droit - Rouge)
ax2 = ax1.twinx()
sns.lineplot(
    x=return_analysis.index.astype(str), 
    y=return_analysis['ReturnRate'], 
    ax=ax2, 
    color='red', 
    marker='o', 
    linewidth=2, 
    label='Taux de Retour (%)'
)

# Config Axe Droit
ax2.set_ylabel('Taux de Retour (% du CA)', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, return_analysis['ReturnRate'].max() * 1.2) # Marge en haut
ax2.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f%%'))

# Titre
plt.title('Suivi des Retours : Montants rembours√©s et Impact sur le CA', fontsize=16, fontweight='bold', pad=20)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Ce graphique nous permet de ne pas piloter √† l'aveugle. Il montre que nous avons un taux de retour moyen de X%. Toute d√©viation au-dessus de cette ligne rouge sera un signal d'alarme pour l'√©quipe produit.

```python

# --- Pr√©paration des donn√©es ---
# On utilise le dataframe 'df_returns' que nous avons isol√© juste avant
# On groupe par Description et on somme le Total (qui repr√©sente le montant rembours√©)
top_returns = df_returns.groupby('Description')['Total'].sum().sort_values(ascending=False).head(10)

# --- Cr√©ation du Graphique ---
plt.figure(figsize=(12, 6))

# Barplot horizontal pour bien lire les noms des produits
# Palette 'Reds_r' pour symboliser la perte/le danger
barplot = sns.barplot(x=top_returns.values, y=top_returns.index, hue=top_returns.index, palette='Reds_r')

# --- Mise en forme ---
plt.title('TOP 10 des Produits g√©n√©rant le plus de pertes (Remboursements)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Montant Total Rembours√© (¬£)', fontsize=12)
plt.ylabel('Produit', fontsize=12)

# Ajout des valeurs au bout des barres pour la pr√©cision
for i, v in enumerate(top_returns.values):
    barplot.text(v + 10, i, f'¬£{v:,.0f}', color='black', va='center', fontweight='bold')

sns.despine()
plt.show()
```
